```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# load a binary classification dataset for breast cancer diagnostic
bc = datasets.load_breast_cancer()
X, y = bc.data, bc.target
```

```python
# split the dataset into the train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)

print("Train set size =", y_train.size)
print("Test set size =", y_test.size)
```

```python
import numpy as np

def sigmoid(x):
  return 1 / (1 + np.exp(-x))

# Class to create and train a logistic regression model
class LogisticRegression:
    def __init__(self, lr = 0.001, n_iters=1000):
        self.lr = lr
        self.n_iters = n_iters
        self.weights = None
        self.bias  = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.n_iters):
            linear_pred = np.dot(X, self.weights) + self.bias
            predictions = sigmoid(linear_pred)

            dw = (2 / n_samples) * np.dot(X.T, (predictions - y))
            db = (2 / n_samples) * np.sum(predictions - y)

            self.weights = self.weights - self.lr * dw
            self.bias = self.bias - self.lr * db

    def predict(self, X, threshold=0.5):
        linear_pred = np.dot(X, self.weights) + self.bias
        y_pred = sigmoid(linear_pred)
        class_pred = [0 if y <= threshold else 1 for y in y_pred]
        return class_pred

    # compute the mean squared error
    def score(self, X, y):
      predictions = self.predict(X)
      return np.sum(predictions == y) / len(y)
```

```python
# train a logistic regression model
lreg = LogisticRegression(lr = 0.02)
lreg.fit(X_train, y_train)
print("Train Accuracy =", lreg.score(X_train, y_train))
print("Test Accuracy =", lreg.score(X_test, y_test))
```

```python
from sklearn.metrics import classification_report

predictions = lreg.predict(X_test)
print(classification_report(y_test, predictions))
```
